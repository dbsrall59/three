```python
import pandas as pd
```

```python
file_path = './행정동코드_매핑정보_2018.xlsx'
xlsx = pd.read_excel(file_path,skiprows=[1],sheet_name='행정동코드',usecols='B:E')
xlsx = xlsx[['행자부행정동코드','행정동명']]
xlsx
```

```python
file_path = './행정동추가.csv'
csv = pd.read_csv(file_path,encoding='euc-kr',sep=',')
csv
```

```python
dong_split = csv['행정동'][0].split(' ')[2]
dong_split
```

```python
t = []
for i in csv['행정동']:
    t.append(i.split(' ')[2])
t    
```

```python
csv['행정동'] = t
csv
```

```python
total = pd.merge(xlsx,csv, how='inner',left_on='행정동명',right_on='행정동')
total = total.drop(['행정동'],axis=1)
total
```

```python
total.to_csv('통행량_동매핑.csv',encoding='euc-kr')
```
