```python
import pandas as pd
```

```python
car = pd.read_csv('c:/users/taihoon/desktop/three/자료/car_불러들이기위한.csv',sep=',')
car
```

```python
# 결측치 처리
car = car.dropna()

# 조건에 맞는 데이터만 추출
car = car[car['연월'] == 202112]

# 인덱스번호 초기화
car = car.reset_index(drop=True)
car
```

```python
# 행정동별로
# 차종별+연료별로 그룹묶어서 퍼센트(%)

# 행정코드, 차종종별, 연료, 퍼센트

# 비율을 내기위한 전체 건수 합
print(car['건수'].sum())

# 그룹지은것 중 건수만 sum 해야 하기에 나머지는 문자열로 변환
car['연월'] = car['연월'].astype(str)
car['행정동사용본거지코드'] = car['행정동사용본거지코드'].astype(str)
car['시군구코드'] = car['시군구코드'].astype(str)

# 필요한 컬럼들만 그룹짓고 합계 구한 후 건수를 기준으로 내림차순 정렬하기
car = (car.groupby(['연월','행정동사용본거지','행정동사용본거지코드','차종종별','연료'],
                        as_index=False).sum()).sort_values(
                            by=['건수'],
                            ascending=False
                        )
car
```

```python
# 백분율 구하기
car['비율'] =  (car['건수']/3172848) *100 

# 인덱스 번호 초기화 하기
car = car.reset_index(drop=True)
car
```
