```python
import pandas as pd
```

```python
car = pd.read_csv('c:/users/taihoon/desktop/three/자료/asd.csv',sep=',')
car['행정동사용본거지코드'] = car['행정동사용본거지코드'].fillna(0)
# car['행정동사용본거지코드'] = pd.to_numeric(car['행정동사용본거지코드'])
car_gun = car[['행정동사용본거지','행정동사용본거지코드','차종종별','연료','건수']]
car_gun
```

```python
#2022년 1월부터 4월까지
# 행정동별로
# 차종별+연료별로 그룹묶어서 퍼센트(%)

# 행정코드, 차종종별, 연료, 퍼센트

car_gun = (car_gun.groupby(['행정동사용본거지','행정동사용본거지코드','차종종별','연료'],
                        as_index=False).sum()).sort_values(
                            by=['건수'],
                            ascending=False
                        )
car_gun
```

```python
car_gun['비율'] =  (car_gun['건수']/21838763) *100 
car_gun = car_gun.reset_index(drop=True)
car_gun
```
