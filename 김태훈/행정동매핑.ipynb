```python
import pandas as pd



file_path = './result.csv'

csv = pd.read_csv(file_path,sep=',')
csv =  csv[csv.columns[1:]]
csv
```

```python
file_path = './행정동코드_매핑정보_2018.xlsx'
xlsx = pd.read_excel(file_path,skiprows=[1],sheet_name='행정동코드',usecols='B:E')
xlsx
```

```python
Dong_result = pd.merge(xlsx,csv,on='행정동코드',how='left')
Dong_result
```

```python
Dong_result.to_csv('Dong_result.csv', sep=',')
```
