```python
#평균재차인원 : 1.22 출처 : https://www.joongang.co.kr/article/22626529#home
carMovingDF = carMovingDF.reindex(columns=['일자', '통행량', '읍면동', '행정동코드', '시군구', '행정구코드'])
carMovingDF['통행량'] = (carMovingDF['통행량'] * 1.22).astype('int')
carMovingDF
```

```python
#carMovingDF 이상치처리
carMovingDF.describe()
```

```python
carMovingDF[carMovingDF['통행량'] == 0]
```

```python
carMeanDF = carMovingDF.groupby(['일자', '시군구'], as_index=False).agg('mean')
carMeanDF
```

```python
#fillna를 대신할 함수
def fillMean(x):
    for i in carMeanDF.index:
        if((x[0] == carMeanDF.iloc[i, 0]) & (x[4] == carMeanDF.iloc[i, 1])):
            x[1] = carMeanDF.iloc[i, 2]
            break;
    return x

carMovingDF[carMovingDF['통행량'] == 0] = carMovingDF[carMovingDF['통행량'] == 0].apply(fillMean ,1)
carMovingDF[carMovingDF['통행량'] == 0]
```

```python
#일별 자치구별 코로나 증가현황
onlyCoronaDF = coronaDF[coronaDF.columns[0::2]]
onlyCoronaDF.sort_values('일자', inplace=True)
onlyCoronaDF
```

```python
#종로구 교통현황 변화 보기
cmJongro = carMovingDF[carMovingDF['시군구'] == '종로구']
chJongro = carHavingDF[carHavingDF['시군구'] == '종로구']
pmJongro = pubilcMovingDF[pubilcMovingDF['시군구'] == '종로구']
```

```python
cmJongroSum = cmJongro.groupby('일자', as_index=False).agg('sum')
pmJongroSum = pmJongro.groupby('일자', as_index=False).agg('sum')

JongroResult = cmJongroSum.merge(pmJongroSum, on='일자')
```

```python
JongroResult['대중교통퍼센트'] = JongroResult['승객수'] /  (JongroResult['통행량'] + JongroResult['승객수']) * 100
JongroResult
```

```python
plt.plot(JongroResult['일자'], JongroResult['대중교통퍼센트'])
plt.xticks(list([x for x in range(0, 111, 5)]), rotation=90)
plt.title('종로구 대중교통 비율')
plt.xlabel('날짜')
plt.ylabel('대중교통비율(%)')
```

```python
#전체 대중교통 비율 보기

allPercent = carMovingDF.merge(pubilcMovingDF, on=['일자', '행정동코드', '행정구코드'])
allPercent
```

```python
allPercent = allPercent[['일자', '통행량', '승객수', '읍면동_x', '행정동코드', '시군구_x', '행정구코드']]
allPercent
```

```python
allPercent.rename({'읍면동_x' : '읍면동', '시군구_x' : '시군구'}, inplace=True, axis=1)
allPercent
```

```python
allPercent['대중교통비율'] = allPercent['승객수'] / (allPercent['통행량'] + allPercent['승객수']) * 100
allPercent
```

```python
#자치구별 대중교통비율

allPercent.info()
guPercent = allPercent.groupby(['일자','시군구'], as_index=False).agg('mean')
guPercent
```

```python
pt = pd.pivot_table(guPercent, index='일자', values='대중교통비율', columns='시군구')
sns.heatmap(pt)
plt.title('일자별 자치구별 대중교통 퍼센트')
```

```python
#자치구별 자동차소유 현황

guHavingGroup = carHavingDF[carHavingDF.columns[:3].append(carHavingDF.columns[4:])].groupby(['차종종별', '연료', '시군구'], as_index=False).agg('sum')
guHavingGroup
```

```python
#필터링
energyFillter = guHavingGroup.sort_values('건수', ascending=False)['연료'].unique()[:6]

def checkEnergy(x):
    if x in energyFillter:
        return True
    return False

guHavingGroup = guHavingGroup[(guHavingGroup['연료'].apply(checkEnergy)) & (guHavingGroup['시군구'] != '0')]
guHavingGroup
```

```python
pt = guHavingGroup.pivot_table(index='연료', values='건수', columns='시군구')
sns.heatmap(pt)
plt.title('자치구별 연료별 자동자등록현황')
```

```python
df3 = df3[df3['차종별']=='계']
df3 = df3.reset_index(drop=True)
df3 = df3.rename(columns={'계':'자동차등록대수(전체)'})
df3
```

```python
df4 = df3[['휘발유','경유','LPG','전기','CNG','하이브리드','수소','기타']]
df4 = df4.T
df4.columns = ['2018','2019','2020','2021']
df4 = df4.reset_index()
df4 = df4.rename(columns={'index':'연료'})
# df4 = df4.T.reset_index()
# df4 = df4.rename(columns={'index':'기간'})
df4
```
