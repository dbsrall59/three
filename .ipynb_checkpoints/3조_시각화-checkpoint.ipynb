```python
pTDF = pd.read_csv(nonSharedDataPath + '행정동별대중교통수단통합.csv', sep=',', encoding='euc-kr')
mappingDF = pd.read_excel(dataPath + '행정동코드_매핑정보_2018.xlsx')
mappingDF = mappingDF[:][1:]
```

```python
mappingDF.info()
pTDF.info()
```

```python
mappingDF[mappingDF.columns[0]] = mappingDF[mappingDF.columns[0]].astype('float')
mappingDF[mappingDF.columns[1]] = mappingDF[mappingDF.columns[1]].astype('float')
```

```python
mappingDF.info()
```

```python
c1 = (pTDF['기준_날짜'] == 0)
c2 = (pTDF['시작_행정동_ID'] == 0)
c3 = (pTDF['종료_행정동_ID'] == 0)
(c1 | c2 | c3).unique()
pTDF = pTDF[(c1 | c2 | c3) == False]
```

```python
for i in pTDF.columns[:3]:
    pTDF[i] = pTDF[i].astype('str')
pTDF.info()
```

```python
shGroup = pTDF.groupby(pTDF.columns[1:3].to_list()).agg('sum').sort_values('승객수', ascending=False)
shGroup = shGroup.head(30)
shGroup
```

```python
plt.rc('font', family='Malgun Gothic', size = 18)
plt.rcParams['figure.figsize'] = (20, 10)
plt.rcParams['lines.linewidth'] = 2

shGroup.plot(xticks=[i for i in range(30)])
plt.xticks(rotation=270)
plt.xlabel('시작행정동 - 종료행정동')
```

```python
carMovingDF = pd.read_csv(dataPath + '행정동추가.csv', sep=',', encoding='euc-kr')
carMovingDF
```

```python
carMovingSumDF = pd.read_csv(dataPath + '통행합계.csv', sep=',', encoding='euc-kr')
carMovingSumDF
```

```python
carMovingSumDF['법정동코드'] = ""
carMovingSumDF['법정동'] = ""
```

```python
for i0 in carMovingSumDF.index:
    for i1 in carMovingDF.index: 
        if carMovingSumDF.loc[i0, '지점명'] == carMovingDF.loc[i1, '지점명칭']:
            carMovingSumDF.loc[i0, '법정동코드'] = carMovingDF.loc[i1, '법정동코드']
            carMovingSumDF.loc[i0, '법정동'] = carMovingDF.loc[i1, '주소'].split(' ')[2]
carMovingSumDF
```

```python
carMovingSumDF[carMovingSumDF['법정동코드'] == ""]['지점명'].value_counts()
```

```python
carMovingSumDF['법정동코드'] = carMovingSumDF['법정동코드'].astype('str')
carMovingSumDF['일자'] = carMovingSumDF['일자'].astype('str')
carMovingSumDF.info()
```

```python
carMovingSumGroup = carMovingSumDF.groupby('지점명').agg('sum')
carMovingSumGroup
```

```python
carMovingSumPivot = pd.pivot_table(carMovingSumDF, index='일자', values='통행량', columns='법정동', aggfunc='sum')
carMovingSumPivot
```

```python
sns.heatmap(carMovingSumPivot)
```
