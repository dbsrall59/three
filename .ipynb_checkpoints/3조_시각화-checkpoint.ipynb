```python
carMovingSumDF = pd.read_csv(dataPath + '통행합계.csv', sep=',', encoding='euc-kr')
carMovingSumDF
```

```python
carMovingSumDF['법정동코드'] = ""
carMovingSumDF['법정동'] = ""
carMovingSumDF['법정구'] = ""
carMovingSumDF['법정구코드'] = ""
```

```python
for i0 in carMovingSumDF.index:
    for i1 in carMovingDF.index: 
        if carMovingSumDF.loc[i0, '지점명'] == carMovingDF.loc[i1, '지점명칭']:
            carMovingSumDF.loc[i0, '법정동코드'] = carMovingDF.loc[i1, '법정동코드']
            carMovingSumDF.loc[i0, '법정동'] = carMovingDF.loc[i1, '행정동']
            carMovingSumDF.loc[i0, '법정구코드'] = carMovingDF.loc[i1, '법정구코드']
            carMovingSumDF.loc[i0, '법정구'] = carMovingDF.loc[i1, '법정구']
carMovingSumDF
```

```python
carMovingSumDF[carMovingSumDF['법정동코드'] == ""]['지점명'].value_counts()
```

```python
carMovingSumDF['법정동코드'] = carMovingSumDF['법정동코드'].astype('str')
carMovingSumDF['일자'] = carMovingSumDF['일자'].astype('str')
carMovingSumDF['법정구코드'] = carMovingSumDF['법정구코드'].astype('str')
carMovingSumDF.info()
```

```python
df3 = pd.read_excel(dataPath + '서울시_자동차_등록현황(연료별)통계.xls')
df3 = df3.fillna(0)
df3
```

```python
df3 = df3[df3['차종별']=='계']
df3 = df3.reset_index(drop=True)
df3 = df3.rename(columns={'계':'자동차등록대수(전체)'})
df3
```

```python
df4 = df3[['휘발유','경유','LPG','전기','CNG','하이브리드','수소','기타']]
df4 = df4.T
df4.columns = ['2018','2019','2020','2021']
# df4 = df4.reset_index()
# df4 = df4.rename(columns={'index':'연료'})
# df4 = df4.T.reset_index()
# df4 = df4.rename(columns={'index':'기간'})
df4
```

```python
import numpy as np

# spot_ = 2018
# d_ = df3[(df3["기간"]==spot_) & (df["일시"].dt.month==9)]
plt.figure(figsize=(15,10))
plt.title("년도별 전기차 등록대수" , fontsize=15)
plt.plot(df3["기간"], df3["전기"], "-", color='blue')
plt.grid()
plt.xticks([2018,2019,2020,2021])
plt.show()
```

```python
# plt.figure(figsize=(15,10))
# plt.title("년도별 전기차 등록대수" , fontsize=15)
# plt.plot(df3["기간"], df3["전기"], "-", color='blue')
# plt.grid()
# plt.xticks([2018,2019,2020,2021])
# plt.show()

plt.figure(figsize=(10,8)) # 먼저 창을 만들고
n=1
color_list = ['red','blue','green','yellow']
for spot_ in ['휘발유','경유','LPG','전기']:
    d_ = df4[(df4["연료"]==spot_)]
    ax = plt.subplot(2,2,n) # for문을 돌면서 Axes를 추가
    plt.title("%s 자동차 등록 현황" % spot_, fontsize=15)
    ax.plot(df3["기간"], df3[spot_], "-", color=color_list[n-1]) # 그래프 추가
    plt.xticks([2018,2019,2020,2021])
    plt.xlabel('년도')
    n+=1
plt.tight_layout()  # 창 크기에 맞게 조정
plt.show()
```
