```python
import pandas as pd
import os
```

```python
df = pd.DataFrame()

for i in range(4, 5):
#     sheetName = '2022년 ' + '{:0>2}'.format(i) + '월'
    sheetName = '수집지점 주소 및 좌표'
    path = 'C:/Users/bigdata/Downloads/서울시교통량/' + '{:0>2}'.format(i) + '월 서울시 교통량 조사자료(2022).xlsx'
    df = pd.concat([df, pd.read_excel(path, sheet_name=sheetName)])
df.fillna('0', inplace=True)
```
