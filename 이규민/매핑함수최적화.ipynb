```python
def binarySearch(mItem, mList):
    
    def bSearch(mItem, mList, startIndex, endIndex):
        
        if(startIndex > endIndex):
            return False
        
        target = int((startIndex + endIndex) / 2)

        if(mItem == mList[target]):
            return target
        elif (mItem < mList[target]):
            return bSearch(mItem, mList, startIndex, target - 1)
        else:
            return bSearch(mItem, mList, target + 1, endIndex)
    
    return bSearch(mItem, mList, 0, len(mList) - 1)

binarySearch(1, [1,2,3,4])
```

```python
def binarySearchWhile(mItem, mList):
    
    startIndex = 0
    endIndex = len(mList) - 1
    
    while(startIndex <= endIndex):
        target = int((startIndex + endIndex) / 2)
        
        if(mItem == mList[target]):
            return target
        elif(mItem < mList[target]):
            endIndex = target - 1
        else:
            startIndex = target + 1
            
    return False

binarySearchWhile(3, [1,2,3,4])
```

```python
mapData = pd.read_csv(dataPath + '행정동추가.csv', encoding='euckr')
mapData = mapData[['행정동']].rename({'행정동' : '읍면동'}, axis=1)
mapData = mapData.sort_values('읍면동').reset_index(drop = True)
mapData
```

```python
basicData = pd.read_excel(dataPath + '행정동코드_매핑정보_2018.xlsx')
basicData = basicData[['통계청행정동코드', '시군구명', '행정동명']][1:]
basicData.rename({'통계청행정동코드' : '행정동코드', '시군구명' : '시군구', '행정동명' : '읍면동'}, inplace=True, axis=1)
basicData = basicData.sort_values('읍면동').reset_index(drop = True)
basicData
```

```python
mapData['행정동코드'] = None
mapData['시군구'] = None

tempSeries = basicData['읍면동'].apply(lambda x : x[:2])

def mappingData(x):
    index = binarySearchWhile(x[0][:2], tempSeries)
    if(index != False):
        x[1] = basicData.iloc[index, 0]
        x[2] = basicData.iloc[index, 1]
    
    return x

mapData = pd.DataFrame(np.apply_along_axis(mappingData, 1, mapData), columns = mapData.columns.to_list())
#mapData = mapData.apply(mappingData, 1)
mapData
```
