```python
import pandas as pd
import os

print(os.getcwd())
```

```python
df = pd.DataFrame()

for i0 in range(1, 5):
    days = 31
    if i0 == 2:
        days = 28
    elif i0 == 4:
        days = 30
        
    for i1 in range(1, days + 1):
        path = 'C:/Users/bigdata/Downloads/tpss_emd_odh_2022' + '{:0>2}'.format(i0) \
        + '/tpss_emd_odh_2022' + '{:0>2}'.format(i0) + '{:0>2}'.format(i1) + '.csv'
        tempdf = pd.read_csv(path, sep=',', encoding='euc-kr')
        df = pd.concat([df, tempdf])
df.fillna(0, inplace=True)
```

```python
df['승객수'] = 0
for i in df.columns[3:27]:
    df['승객수'] += df[i]
df
```

```python
df[['기준_날짜', '시작_행정동_ID', '종료_행정동_ID', '승객수']]\
.to_csv('행정동별대중교통수단통합.csv', sep=',', encoding='euc-kr', index=False)
df = pd.read_csv('행정동별대중교통수단통합.csv', sep=',', encoding='euc-kr')
df
```
