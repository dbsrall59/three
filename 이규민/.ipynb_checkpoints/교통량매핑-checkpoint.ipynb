```python
df2 = pd.read_csv('법정동코드 전체자료.csv', sep=',', encoding='euc-kr')
df2
```

```python
guList = []
for i in df2.index:
    temp = df2.loc[i, '법정동명'].split()
    if(len(temp) == 2) & (temp[0] == '서울특별시'):
        guList.append(i)
guList
```

```python
c1 = []
c2 = []
c3 = []

guIndex = 0

for i in df2.index:
    temp = df2.loc[i, '법정동명'].split(' ')
    
    if(i == guList[guIndex]):
        guIndex += 1
        
    if(len(temp) < 3):
        c1.append("")
        c2.append("")
        c3.append("")
        
    else:
        c1.append(temp[2])
        c2.append(temp[1])
        c3.append(df2.loc[guList[guIndex - 1], '법정동코드'])
    
df2['법정동'] = c1
df2['법정구'] = c2
df2['법정구코드'] = c3

df2
```

```python
df.loc[guList[3], '법정동코드']
```

```python
df['법정동코드'] = ""
for i0 in df.index:
    for i1 in range(1112):
        if df.loc[i0, '행정동'] == df2.loc[i1, '법정동']:
            df.loc[i0, '법정동코드'] = df2.loc[i1, '법정동코드']
df
```
