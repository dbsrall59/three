```python
df2 = pd.read_csv('법정동코드 전체자료.csv', sep=',', encoding='euc-kr')
df2
```

```python
guList = []
for i in df2.index:
    temp = df2.loc[i, '법정동명'].split()
    if(len(temp) == 2) & (temp[0] == '서울특별시'):
        guList.append(i)
guList
```

```python
c1 = []
c2 = []
c3 = []

guIndex = 0
for i in df2.index:
    #존재없는거 뺴기
    temp = df2.loc[i, '법정동명'].split(' ')
        
    for i1 in range(len(guList)):
        if(guList[i1] == i):
            guIndex = i1
            break
        
    if(len(temp) < 3):
        c1.append("0")
        c2.append("0")
        c3.append("0")
        
    else:
        c1.append(temp[2])
        c2.append(temp[1])
        c3.append(df2.loc[guList[guIndex], '법정동코드'])
    
df2['법정동'] = c1
df2['법정구'] = c2
df2['법정구코드'] = c3

df2 = df2[:][:1112]
```

```python
df2.to_csv('법정코드매핑.csv', sep=',', encoding='euc-kr', index=False)
```
